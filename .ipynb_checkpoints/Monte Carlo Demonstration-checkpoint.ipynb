```python
import numpy as np
import progressbar
import matplotlib.pyplot as plt
from math import pi
```

```python
def test_point():
    pt = np.random.rand(2)
    distance_from_center = np.linalg.norm(pt - np.array([0.5, 0.5]))
    return (pt, 1) if distance_from_center <= 0.5 else (pt, 0)
```

```python
def simulate(n, k, l):
    tests = []
    approximations = []
    points = []
    pt_exports = []
    for i in progressbar.progressbar(range(n)):
        pt, test = test_point()
        points.append(pt)
        tests.append(test)
        if i%k==0 and i>0:
            approximations.append(np.mean(np.array(tests)) * 4.0)
        if i+1 in l:
            pt_exports.append(points)
    return approximations, pt_exports
```

```python
approximations, points = simulate(100000, 10, [10, 100, 1000, 10000, 100000])
```

```python
approx = np.array(approximations)
pct_diff = np.abs((pi-approx)/pi)
```

```python
plt.figure(figsize=(16, 9), dpi=80)
plt.scatter(np.arange(len(pct_diff)), pct_diff)
plt.title("Monte Carlo Pi Approximation")
plt.xlabel("Iterations")
plt.ylabel("Error %")
plt.show()
```
